```python
import pandas as pd
import os
```

```python
data_folder = "../data/raw/tls/data.xlsx"

# load data
data_df = pd.read_excel(data_folder, engine="openpyxl", sheet_name=0)
data_df.head()
```

```python
data_df = data_df[["Sample", "Intent"]]
data_df.sample(n=10, random_state=42)
```

```python
data_df["Sample"] = data_df["Sample"].apply(str.lower) # lower text
data_df.sample(n=10, random_state=42)
```

```python
classes = set(data_df["Intent"].tolist())
classes
```

```python
X_dict = {utterance: [] for utterance in data_df["Sample"].tolist()}

if len(X_dict) < len(data_df.index):
    print("Multilabel Dataset")
else:
    print("Non Multilabel Dataset")
```

```python
for _, pack in data_df.iterrows():
    utterance = pack["Sample"]
    intent = pack["Intent"]
    X_dict[utterance].append(intent)
```

```python
list(X_dict.keys())
```

```python
X = list(X_dict.keys())
y = list(X_dict.values())

processed_path = "../data/processed/tls/"
X_path = os.path.join(processed_path, "text")
y_path = os.path.join(processed_path, "label")

def write_data(path, data, is_label=False):
    with open(path, "w") as f:
        for e in data:
            e = ",".join(e) if is_label else e
            f.write(e)
            f.write("\n")

write_data(X_path, X)
write_data(y_path, y, is_label=True)
```
